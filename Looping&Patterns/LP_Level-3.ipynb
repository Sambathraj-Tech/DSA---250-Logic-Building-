1. Print the squares of numbers from 1 to n.

```python
def squares(n):   
    for i in range(1, n+1, 1):
        print(i * i , end= " ")
squares(6)
print()
squares(5) 
```

2. Print Cubes from 1 to n

```python
import math
def cubes(n):   
    for i in range(1, n+1, 1):
        print(pow(i, 3), end= " ")
cubes(6)
print()
cubes(5) 
```

3. Print all numbers between a and b divisible by 7.

```python
def divisible(n):
    for i in range(1, n + 1, 1):
        if n % i == 0:
            print(i, end= " ")

divisible(7)
print()
divisible(54)
print()
divisible(30)
```

4. Find HCF of two numbers using loops.

```python
def hcf(a, b):
    largest = 1
    for i in range(2, min(a, b)):
        if a % i == 0 and b % i == 0:
            largest = i
    return largest

print(hcf(4, 6))
print(hcf(25, 60))
print(hcf(5, 10))
print(hcf(9, 8))
```

5. LCM of two numbers using loops.

```python
def LCM(a, b):
    lcm = 0
    n = max(a, b)
    i = 1
    while True:
        mult = n * i
        if mult % a == 0 and mult % b == 0:
            lcm = mult
            break
        i += 1
    return lcm
print(LCM(10, 5))
print(LCM(12, 8))
print(LCM(6, 4))
```

6. Print all factors of a given number.

```python
import math
def Factors(n):
    if n == 1:
        return 1
    result = []
    for i in range(2, int(math.sqrt(n) + 1), 1):
        if n % i == 0:
            result.append(i)
            if i != (n // i):
                result.append(n // i)
    result.sort()
    return result
print(Factors(12))
print(Factors(6))
print(Factors(24))
```

7. Find the sum of all factors of a number.

```python
import math
def sum_factors(n):
    result = 0
    for i in range(1, int(math.sqrt(n)) + 1, 1):
        if n % i == 0:
            result += i
            if (n // i) != i:
                result += (n // i)
    return result
print(sum_factors(20))
print(sum_factors(6))
print(sum_factors(1))
print(sum_factors(12))
print(sum_factors(28))
print(sum_factors(7))
```

8. Check if a numebr is Strong number. (sum of factorials of digits = number)

```python
class solution:
    def factorial(self, n): # factorial function
        fact = 1
        for i in range(1, n + 1):
            fact = fact * i
        return fact
    def is_factorial(self, n): # Values passing function
        copy = n
        result = 0
        while n > 0:
            temp = n % 10
            result += self.factorial(temp)
            n = n // 10
        return copy == result # Check input == result

sol = solution()
print(sol.is_factorial(123))
print(sol.is_factorial(1))
print(sol.is_factorial(145))
print(sol.is_factorial(2))
print(sol.is_factorial(40585))
print(sol.is_factorial(10))
```

9. Print first n terms of an arithmetic progression(a,d). ==>  a ,a+d, a+2d, a+3d.....

```python
def arithmetic_progression(a, d, n):
    for i in range(1, n+1, 1):
        result = a + (i-1) * d
        print(result, end=" ")

arithmetic_progression(2, 3, 5)
print()
arithmetic_progression(5, 2, 7)
print()
arithmetic_progression(10, -3, 6)
```

10. Print first n terms of an geometric progression(a,r). (a + r * (n-1))

```python
def geometric_progression(a, r, n):
    for i in range(1, n + 1, 1):
        result = a + r * (i - 1)
        print(result, end=" ")
geometric_progression(2, 3, 5) 
print()
geometric_progression(5, 2, 6)
print()
geometric_progression(64, 0.5, 7)
```
