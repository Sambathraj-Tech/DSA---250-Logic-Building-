Level - 1 : Basic Looping

1. Print numbers from 1 to 10

```python
def numbers(n):
    for i in range(1, n + 1, 1):
        print(i, end=" ")

numbers(10)
```

2. Print all even numebrs from 1 and 100.

```python
def even_numbers(n):
    for i in range(1, n+1, 1):
        if i % 2 == 0:
            print(i, end=" ")

even_numbers(100)
```

3. Print all odd numbers from 1 and 100.

```python
def odd_numbers(n):
    for i in range(1, n+1, 1):
        if i % 2 == 1:
            print(i, end=" ")

odd_numbers(100)
```

4. Print numbers from down 10 to 1.

```python
def number(n):
    for i in range(n, 0, -1):
        print(i, end=" ")
number(10)
```

5. print the tables of a given number (n * 1 to n * 10).

```python
def tables(n):
    for i in range(1, 10+1, 1):
        ans = n * i
        print(f"{n} * {i} = {ans}")
tables(2)
```

6. Print the sum of first n natural numbers.

```python
def natural(n):
    total = 0
    for i in range(1, n+1, 1):
        total += i
    return total

print(natural(5))
print(natural(3))
```

7. Print the sum of all even numbers up to n.

```python
def even(n):
    total = 0
    for i in range(1, n+1, 1):
        if i % 2== 0:
            total += i
    return total

print(even(10))
```

8. Print the sum of all odd numbers up to n.

```python
def odd(n):
    total = 0
    for i in range(1, n+1, 1):
        if i % 2 == 1:
            total += i
    return total

print(odd(10))
```

9. Print the factorial of the given number.

```python
def factorial(n):
    fact = 1
    for i in range(1, n+1, 1):
        fact *= i
    return fact

print(factorial(10))
print(factorial(5))
print(factorial(3))
```

10. Print the product of digits of a given number.

```python
def product(n):
    result = 1
    while n > 0:
        result *= (n % 10)
        n = n // 10
    return result 

print(product(123))
print(product(14234))
print(product(456))
```
