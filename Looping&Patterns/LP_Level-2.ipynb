Level - 2: Number Based Looping Logic.
1. Count the number of digits in a given number.

```python
def counted(n):
    count = 0
    while n > 0:
        count += 1
        n = n // 10
    return count
print(counted(123))
print(counted(123223))
print(counted(123294785))
```

2. Print reverse the given number.

```python
def reverse(n):
    result = 0
    while n > 0:
        temp = n % 10
        result = (result * 10) + temp
        n = n // 10
    return result
print(reverse(123))
print(reverse(123223))
print(reverse(123294785))
```

3. Chech if a number is palindrome or not.

```python
def reverse(n):
    result = 0
    copy = n
    while n > 0:
        temp = n % 10
        result = (result * 10) + temp
        n = n // 10
    if result == copy:
        return "Palindrome"
    else:
        return "Not palindrome"

print(reverse(123))
print(reverse(12221))
print(reverse(123294785))
```

4. Find the sum of digits of a number.

```python
def total(n):
    result = 0
    while n > 0:
        temp = n % 10
        result += temp
        n = n // 10
    return result 
print(total(72345))
print(total(45))
print(total(723))
```

5. Check if a number is Armstrong or not.

```python
def armstrong(n):
    copy = n
    result = 0
    length = len(str(n))
    while n > 0:
        temp = n % 10
        result = result + (temp ** length)
        n = n // 10
    return "Armstrong number" if (copy == result) else "Not Armstrong number"
print(armstrong(153))
print(armstrong(30))
print(armstrong(1634))
```

6. Check if a numebr is a perfect number.

```python
import math
def perfect(n):
    if n <= 0:
        return False
    result = 0
    for i in range(1, int(math.sqrt(n) + 1), 1):
        if n % i == 0:
            result += i
            if i != n // i and n // i != n:
                result += n // i
    return result

print(perfect(6))
print(perfect(36))
print(perfect(12))
```

7. Print all prime numbers between 1 and 100.

```python
import math
def isprime(n):
    if n <= 1 :
        return False
    elif n == 2:
        return True
    if n % 2 == 0:
        return False # Even numbers cannot be a prime
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True

def prime(n):
    result = []
    for n in range(1, 100+1):
        if isprime(n) == True:
            result.append(n)
    return result
print(prime(100), end= " ")
```

8. Check Prime or not

```python
import math
def isprime(n):
    if n <= 1 :
        return False
    elif n == 2:
        return True
    if n % 2 == 0:
        return False # Even numbers cannot be a prime
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True

def prime(n):
    if isprime(n) == True:
        return "It is a prime number", n
    return "It is not a prime number", n

print(prime(7))
print(prime(23))
print(prime(36))
print(prime(56))
print(prime(17))
```

10. Fibonacci series sum of first n terms.

```python
def fibonacci(n):
    a , b = 0, 1
    result = 0
    for i in range(1, n+1, 1):
        result += a
        a, b = b, a + b
    return result
print(fibonacci(20))
print(fibonacci(0))
print(fibonacci(7))
print(fibonacci(1))
print(fibonacci(2))
```
