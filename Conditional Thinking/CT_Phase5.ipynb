1. Take coordinates (x, y) and check if the point lies on the X-axis , Y-axis or at the origin.
    If x = 0 and y = 0, the point lies at the Origin.
    If x = 0 and y ≠ 0, the point lies on the Y-axis.
    If x ≠ 0 and y = 0, the point lies on the X-axis.
    Otherwise, it lies in none of these categories.

```python
def coordinates(x, y):
    if x == 0 and y == 0:
        return "Origin"
    elif x == 0:
        return "Lies on the Y-axis"
    elif y == 0:
        return "Lies on the X-axis"
    else:
        return "None"

print(coordinates(0, 0))
print(coordinates(9, 0))
print(coordinates(0, 8))
print(coordinates(2, 8))
```

2. Take three numbers and check if they can perform Pythagorean triplet.
A set of three numbers forms a Pythagorean triplet if:

    (largest number)^2 = (other number)^2 + (other number)^2 

In other words, if the square of the largest value equals the sum of squares of the other two values.
Print "YES" if they form a Pythagorean triplet, otherwise print "NO".

```python
def pythagorean(num1, num2, num3):
    if num1 <= 0 or num2 <= 0 or num3 <= 0:
        return "Invalid Input"
    elif (num1 * num1) + (num2 * num2) == num3 * num3:
        return "num3 is greatest"
    elif (num3 * num3) + (num2 * num2) == num1 * num1:
        return "num1 is greatest"
    elif (num1 * num1) + (num3 * num3) == num2 * num2:
        return "num2 is greatest"
    else:
        return "False"

print(pythagorean(5, 4, 3))
print(pythagorean(6, 8, 10))
print(pythagorean(2, 3, 4))
```

Take time (hours and minutes) and print the smaller angle between the hour and minute hands.
Important Details / Formula:

    * Convert hours to 12-hour format: h = hours % 12.
    * Minute hand angle from 12 o’clock: minute_angle = minutes × 6 degrees (since 360° / 60 = 6° per minute).
    * Hour hand angle from 12 o’clock: hour_angle = h × 30 + minutes × 0.5 degrees (30° per hour + 0.5° per minute).
    * Absolute difference: angle = |hour_angle − minute_angle|.
    * Smaller angle: result = min(angle, 360 − angle).

```python
import math
def angle_degree(hour, minutes):
    if not (1 <= hour <= 24 and 0 <= minutes <= 59):
        return "Not valid "

    h = hour % 12
    minute_angle = minutes * 6
    hour_angle   = (h * 30) + (minutes * 0.5)
    angle = abs(hour_angle - minute_angle) 
    result = min(angle, 360 - angle)
    return result

print(angle_degree(3, 00))
print(angle_degree(12, 30))
print(angle_degree(9, 15))
print(angle_degree(23, 59))
print(angle_degree(26, 59))
```

5. Take three number and check if they are arithmetic progression.(b−a=c−b)

```python

def arithmetic_progression(a, b, c):

    num1 = b - a
    num2 = c - b
    return num1 == num2
    
print(arithmetic_progression(10, 20, 30))
print(arithmetic_progression(3, 5, 7))
print(arithmetic_progression(35, -9, -9))
```

6. Take three number and check if they are Geometric progression

```python
def geometric_progression(a, b, c):

    num1 = b // a
    num2 = c // b
    return num1 == num2
    
print(geometric_progression(2, 6, 18))
print(geometric_progression(3, 5, 7))
print(geometric_progression(35, -9, -9))
```

8. Take an integer (1 - 9999) and check if the sum of its digits is greater than the product of its digits.

```python
def sum_product(n):
    if not (1 <= n <= 9999):
        return "Invalid Input"

    total_sum = 0
    total_pro = 1
    while n > 0:
        temp = n % 10
        total_sum += temp
        total_pro *= temp
        n = n // 10
    return total_sum > total_pro

print(sum_product(-123))
print(sum_product(1234))
print(sum_product(1112))
```

9. Take two dates (day and month) and determine which one comes first in the calendar.

```python
def calendar(date1, month1, date2, month2):
    if date1 == date2 and month1 == month2:
        return "Identical"
    elif month1 < month2:
        return f"Smallest :date1"
    elif month1 == month2 and date1 < date2:
        return f"Smallest : date1"
    else:
        return f"Smallest :date2"

print(calendar(5, 3, 12, 2))    
print(calendar(10, 10, 10, 10))    
print(calendar(25, 12, 1, 1))    
print(calendar(25, 1, 1, 12))    
    
```
