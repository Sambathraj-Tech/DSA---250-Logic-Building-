1. Take three slides and check if they form a valid triangle.
Problem: Triangle Validity Check

Input:
A single line with three positive numbers (integers or decimals) separated by spaces.
Example:
3 4 5

Output:
Print YES if the three lengths can form a valid triangle, otherwise print NO.

Examples
Input: 1 2 3
Output: NO

Input: 5 5 8
Output: YES

```python
def validity(a, b, c):
    if (a <= 0 or b <= 0 or c <= 0):
        return "False"
    elif(a + b > c and b + c > a and c + a > b):
        return "True"    
    return "False"

a, b, c = 2, 34, 21
print(validity(a, b, c))
```

2. If the slides forms valid triangle, determine whether its equilateral, isosceles or scalene. 

Examples
Input:3 3 3
Output: EQUILATERAL

Input:4 4 6
Output:ISOSCELES

Input:3 4 5
Output:SCALENE

Input:1 2 3
Output:INVALID

```python
def validity(a, b, c):
    if (a <= 0 or b <= 0 or c <= 0):
        return "Enter valid inetegers"

    if not (a + b > c and b + c > a and c + a > b):
        return "This is not a Valid triangle."
    elif a == b == c:
        return "This is Equilateral triangle."
    elif a == b or b == c or c == a:
        return "This is Isosceles triangle."
    else:
        return "This is Scalene triangle."  
    

a, b, c = 23, 34, 21
print(validity(a, b, c))
```

3. Take marks (0 - 100) and print the corresponding Grade.(A/B/C/D/F)

```python
def grade(marks):
    if (marks > 100) or (marks < 0):
        return "Invalid marks" 

    elif 90 <= marks <= 100 :
        return "Grade A"
    elif marks >= 70:
        return "Grade B"
    elif marks >= 50:
        return "Grade C"
    elif marks >= 35:
        return "Grade D"
    elif 0 <= marks:
        return "Grade F"

    
marks = 105
print(grade(marks))
```

5. Take the hour of the day (0–23) and print “Good Morning”,“Good Afternoon”,“GoodEvening”or “Good Night”

```python
def clock(hour):
    if 0 <= hour <= 11:
        return "Good Morning"
    elif hour <= 15:
        return "Good Afternoon"
    elif hour <= 20:
        return 'Good Evening'
    elif hour < 24:
        return "Good Night"
    else:
        return "Time is Invalid"
    
hour = 23.50
clock(hour)
```

7. Take two numbers and determine whether both are even, both are odd, or one is even and one is odd.

```python
def odd_even(num1, num2):
    if num1 % 2 == 0 and num2 % 2 == 0:
        print("Both are even.")
    elif num1 % 2 != 0 and num2 % 2 != 0:
        print("Both are odd.")
    elif num1 % 2 != 0 and num2 % 2 == 0:
        print(f"{num1} is Odd, {num2} is Even")
    elif num1 % 2 == 0 and num2 % 2 != 0:
        print(f"{num1} is Even, {num2} is Odd")
    else:
        return "Invalid Input"
    
odd_even(12, 34)
```

8. Take an alphabet character and check if it lies between ‘a’ and ‘m’ or ‘n’ and ‘z’.

```python
def alphabet(s):
    if len(s) != 1 or not s.isalpha():
        return "Invalid Input"
    
    s = s.lower()
    if 'a' <= s <= 'm':
        print("It lies between a-m.")
    else:
        print("It lies between n-z.")
    
s = '#'
alphabet(s)
```
