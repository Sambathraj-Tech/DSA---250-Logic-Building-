1. Take a 3 digit number and check all the digits are Distinct.
Input:A single 3-digit integer.
Example:352

Output:
Print one of the following:
DISTINCT — if all digits are different
NOT DISTINCT — if any digit repeats

```python
digits = 352
check = len(str(digits))
result = set()
while digits > 0:
    temp = digits % 10
    result.add(temp)
    digits = digits // 10

print(result)
if len(result) == check:
    print('True')
```

2. Take three digit number and determine if the middle digit is the largest , smallest or neither.

You need to compare the middle digit with the other two digits:

Is it the largest of all three?
Is it the smallest of all three?
Or is it neither (somewhere in between)?

Examples:

152 → Middle digit is 5. Compare: 1, 5, 2 → 5 is largest ✅
951 → Middle digit is 5. Compare: 9, 5, 1 → 5 is neither (in between)
321 → Middle digit is 2. Compare: 3, 2, 1 → 2 is neither (in between)
123 → Middle digit is 2. Compare: 1, 2, 3 → 2 is neither (in between)
512 → Middle digit is 1. Compare: 5, 1, 2 → 1 is smallest ✅
555 → Middle digit is 5. Compare: 5, 5, 5 → All same, what's the answer?

```python
def lsn(number):
    if number < 100 or number > 999:
        return "Invalid Input"
    
    
    first_digit  = (number // 100)
    middle_digit = (number // 10) % 10
    last_digit   = number % 10

    if middle_digit == first_digit == last_digit:
        return "All are Equal"
    elif middle_digit > first_digit and middle_digit > last_digit:
        return "Middle digit is largest"
    elif middle_digit < first_digit and middle_digit < last_digit:
        return "Middle digit is smallest"
    else:
        return "Neither"

number = 363
print(lsn(number))
```

6. Take coordinates of (x, y) and determine which Quadrant the point lies.

```python
def axis(x, y):
    if x == 0 and  y == 0:
        return "Orgin"
    elif x == 0:
        return "On x axis"
    elif y == 0:
        return "On y axis"
    elif x > 0 and y > 0:
        return "Quadrant - 1"
    elif x < 0 and y > 0:
        return "Quadrant - 2"
    elif x < 0 and y < 0:
        return "Quadrant - 3"
    elif x > 0 and y < 0:
        return "Quadrant - 4"

print(axis(0, 0))
print(axis(0, 2))
print(axis(1, 0))
print(axis(1, 2))
print(axis(-1, 2))
print(axis(-1, -2))
print(axis(1, -2))
```

7. Check if an amount can be evenly divided into 2000, 500 and 100 currency notes.

```python
def divide(number):
    if number < 100 or number == 100:
        return "Invalid Input"
    elif number % 100 == 0:
        return "It is divisble by currency notes"
    else:
        return "Not Possible"

print(divide(100))      
print(divide(87))
print(divide(1250))      
print(divide(1000))
print(divide(7500))       
print(divide(2000)) 
```

8. Check if a number lies within the range [100, 999].

```python
def f(n):
    if 100 <= n <= 999:
        return f"{n} lies in the range [100 - 900]"
    else:
        return "It doesn't lies in the range [100 - 900]"

print(f(1))
print(f(100))
print(f(-100))
print(f(341))
print(f(1000))
```

9. Take two angles of a triangle and compute the third angle.

```python
def triangle(a, b):
    if (a + b) >= 180:
        return "Triangle is invalid because the two angle exeecdes 180"
    elif a == 0 or b == 0:
        return "Invalid input"

    c = 180 - (a + b)
    return c

print(triangle(0, 80))
print(triangle(60, 80))
print(triangle(45, 45))
print(triangle(90, 30))
print(triangle(90, 95))
```

10.Check whether a number is a perfect square (without using the square root function).

Perfect squares:
1. 0     → Yes (0 × 0) ✅              input(num)     output
2. 1     → Yes (1 × 1) ✅                 25 ✅        5 * 5 = 25 
3. 4     → Yes (2 × 2) ✅                 36 ✅        6 * 6 = 36
4. 9     → Yes (3 × 3) ✅                125 ❌        10 * 10 = 100
5. 16    → Yes (4 × 4) ✅                              11 * 11 = 121
6. 25    → Yes (5 × 5) ✅
7. 100   → Yes (10 × 10) ✅
8. 144   → Yes (12 × 12) ✅
9. 10000 → Yes (100 × 100) ✅

Not perfect squares:
10. 2    → No ❌
11. 3    → No ❌
12. 5    → No ❌
13. 7    → No ❌
14. 10   → No ❌
15. 15   → No ❌
16. 99   → No ❌

Invalid:
17. -4   → No (negative) ❌
18. -16  → No (negative) ❌

```python
def perferct(num):
    if num < 0:
        return "Negative numbers not accpetable."
    elif num == 0:
        return "0 is a perfect square."
    
    i = 1
    while i * i <= num:
        if i * i == num:
            return "Its a perfect square"
        i += 1
    else:
        return "Its not perfect square." 


print(perferct(25))
print(perferct(15))
print(perferct(36))
print(perferct(26))
```
