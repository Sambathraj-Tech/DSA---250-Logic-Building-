1. Take a character and check if it is a letter, a digit, or neither.

```python
def check(ch):
    if len(ch) != 1:
        return "Invlaid Input"
    elif 'a' <= ch <= 'z' or 'A' <= ch <= 'Z':
        return "Letter"
    elif '0' <= ch <= '9':
        return "Digit"
    else:
        return "Neither"

print(check('a'))
print(check('3'))
print(check('@'))
print(check('we'))
```

2. Problem Statement:
Given an integer n, print a string based on the following rules:
    Print "Fizz" if n is divisible by 3.
    Print "Buzz" if n is divisible by 5.
    Print "FizzBuzz" if n is divisible by both 3 and 5.
    If n is divisible by neither 3 nor 5, print the number itself as a string.

```python
def func(n):
    if n % 3 == 0 and n % 5 == 0:
        return "FizzBuzz"
    elif n % 3 == 0:
        return "Fizz"
    elif n % 5 == 0:
        return "Buzz"
    else:
        return f"{n}"
    
print(func(15))
print(func(25))
print(func(9))
print(func(23))
print(func(5))
```

3. Problem Statement:
Given three distinct integers a, b, and c, find and print the median value, i.e.,
the number that is neither the maximum nor the minimum among the three.

Example 1: Numbers are 5, 2, 8. Min = 2, Max = 8, Median = 5.
Example 2: Numbers are 10, 10, 5. Min = 5, Max = 10, Median = 10 (since two numbers are equal).
Example 3: Numbers are 7, 3, 7. Min = 3, Max = 7, Median = 7.

```python
def median(a, b, c):
    if a == b or a == c : 
        return a
    elif b == c:
        return b
    elif (a > b) and (a < c) or (a > c) and (a < b): # Main Logic Write in note
        return a
    elif (b > a) and ( b < c) or (b < a) and (b > c):
        return b
    else:
        return c
    

def median2(a, b, c):
    return a + b + c - (max(a, b, c) + min(a, b , c))


print(median(2, 5, 8) , end =" ")
print(median(10, 10, 8), end =" ")
print(median(5, 7, 7))

print(median2(2, 5, 8), end =" ")
print(median2(10, 10, 8), end =" ")
print(median2(5, 7, 7), end =" ")
```

4. Given the hours and minutes of a time in 24-hour format, print whether the time is in AM or PM.
    The input hours will be in the range 0–23.
    Minutes will be in the range 0–59.
    The output should print "AM" if the time is before noon (00:00 to 11:59) and "PM" if the time is noon or later (12:00 to 23:59).
    This problem tests your understanding of conditional logic, time formatting, and basic comparison operations.

```python
def time(hours, minutes):
    if not (0 <= hours <= 23 and 0 <= minutes <= 59):
        return "Invalid Input"
    elif hours < 12:
        return "AM"
    else:
        return "PM"

print(time(9, 30))
print(time(14, 45))
print(time(12, 00))
print(time(00, 15))
print(time(10, 90))
print(time(25, 10))
```

5. Problem Statement:
You are given two values:
    an integer age, and
    an integer income (in lakhs per year).

Your task is to determine whether the person is eligible to pay tax based on the following rules:
    A person is eligible for tax if:
        age > 18, and
        income > 5 lakhs

If both conditions are satisfied, print "Eligible".
Otherwise, print "Not Eligible".

```python
def eligible(age, income):
    if not (0 <= age <= 120) or (income < 0):
        return "Invalid Input"
    elif age > 18 and income > 5:
        return "Eligible"
    else:
        return "Not Eligible"
    
print(eligible(19, 5.0))
print(eligible(18, 10))
print(eligible(19, 6))
print(eligible(19, -6))
```

6. Take two numbers and check if both are positive and their sum is less than 100.

```python
def total(num1 ,num2):
    if (num1 <= 0 or num2 <= 0):
        return "Invalid input"
    total_sum = num1 + num2
    if total_sum <= 100:
        return f"Both are Positive and their sum is {total_sum}"
    else:
        return "Sum is Grater than 100"

print(total(20, 30))
print(total(60, 50))
print(total(-5, 10))
print(total(30, 0))
```

7. Take a Single Digit (0-9) and print its word from ("Zero to Nine)

```python
number_names = ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
def word(num):
    if not isinstance(num, int):
        return "Invalid Input for Decimal"
    elif (0 <= num <= 9):
        return number_names[num]    
    else:
        return "Invalid Input"
    
print(word(9))
print(word(0))
print(word(10))
print(word(5.5))
```

8. Take a Week day number (1 - 7) and determine if it is a weekday or weekend

```python
def Week(day):
    if not isinstance(day, int):
        return "Not valid"
    elif day not in range(1, 7+1):
        return "Invalid Input"
    elif day in (1, 7):
        return "Weekend"
    else:
        return "Weekday"
    
print(Week(3))
print(Week(6))
print(Week(7))
print(Week(9))
```

9. Take Electricity units consumed and calculate the bills as per slabs(using if-else).
| Units Range | Rate per Unit |
| ----------- | ------------- |
| 0 – 100     | ₹5 per unit   |
| 101 – 200   | ₹7 per unit   |
| 201 – 300   | ₹10 per unit  |
| Above 300   | ₹12 per unit  |

```python
def electricity_bill(unit):
    if not isinstance(unit, int):
        return "Invalid input"  
    elif unit < 0:
        return "Invalid Input"
    elif unit <= 100:
        return f"Total amount = {unit * 5}"
    elif unit <= 200:
        return f"Total amount = {unit * 7}"
    elif unit <= 300:
        return f"Total amount = {unit * 10}"
    else:
        return f"Total amount = {unit * 12}"

print(electricity_bill(50))
print(electricity_bill(150))
print(electricity_bill(250))
print(electricity_bill(350))
print(electricity_bill(-50))
print(electricity_bill(50.80))
```

10. Take a Password string and check the basic rules (length >= 8 contains at least one digit).

```python
def password_checker(password):
    length = len(password)
    if (length <= 8):
        return "Password must contain 8 characters"
    has_digit = any(char.isdigit() for char in password)
    if not has_digit:
        return "Password must contain one digits"
    
    return 'Valid Password'


print(password_checker('hellasdfo123'))
print(password_checker('A1b!'))
print(password_checker('abcde'))
```
